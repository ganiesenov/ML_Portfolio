```python
import pandas as pd 
import numpy as np 
```

```python
from sklearn import datasets


X_disbalance, y_disbalance = datasets.make_classification(
    n_samples=20,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_classes=2, 
    random_state=2,
    weights=(0.9, 0.1)
)

X_disbalance, y_disbalance
```

```python
pred = [0 for i in range(20)]
```

```python
pred_df_disbalance = pd.DataFrame({
    'ground truth': y_disbalance,
    'prediction': pred
})

pred_df_disbalance
```

```python
pred_df_disbalance['right']  = pred_df_disbalance['ground truth'] == pred_df_disbalance['prediction']
pred_df_disbalance
```

```python
without_pers = pred_df_disbalance['right'].sum() / pred_df_disbalance['right'].shape[0]
```
